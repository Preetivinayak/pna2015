```python
import nibabel as nib
img = nib.load('ds114_sub009_t2r1.nii')
data = img.get_data()
```

```python
from scipy.optimize import fmin_powell
from scipy.ndimage import affine_transform
```

```python
fixed_data = data[..., 1:]
```

```python
run_mean = fixed_data.mean(axis=-1)
run_mean.shape
```

```python
plt.imshow(run_mean[..., 17])
```

```python
def cost_function(trans_rots):
    trans = trans_rots[:3]
    M = nib.eulerangles.euler2mat(*trans_rots[3:])
    out_vol = affine_transform(run_mean, M, trans, order=1)
    return mismatch(out_vol, single_vol)
```

```python
def mismatch(X, Y):
    return -np.corrcoef(X.ravel(), Y.ravel())[0, 1]
```

```python
def my_callback(params):
   print("Trying parameters " + str(params))
```

```python
single_vol = fixed_data[..., 0]
best_params = fmin_powell(cost_function, np.zeros(6), callback=my_callback)
best_params
```

```python
single_vol = fixed_data[..., -1]
best_params = fmin_powell(cost_function, np.zeros(6), callback=my_callback)
best_params
```

```python
from nipy.algorithms.registration import HistogramRegistration
```

```python
to_img = nib.Nifti1Image(run_mean, img.affine, img.header)
from_img = nib.Nifti1Image(data[..., -1], img.affine, img.header)
R = HistogramRegistration(from_img, to_img, similarity='cc', interp='tri')
```

```python
res = R.optimize('rigid', optimizer='powell')
```
